```python
# Stack Implementation using singly linked List
class Node:
    def __init__(self,val):
        self.id = val
        self.next = None    

class Stack:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def push(self,val): 
        """ Time Complexity:o(1)"""
        if(self.head is None):
            self.head = Node(val)  # O(1)
            self.tail = self.head  # O(1)
            #  O(1) + O(1) = O(1)
        else:
            self.tail.next = Node(val) # O(1)
            self.tail = self.tail.next # O(1)
            #  O(1) + O(1 =O(1)
        # O(1) or O(1) = O(1)
    
        
    def top(self):
        """ Time Complexity:O(1)"""
        return self.tail.id  # O(1)
    
    def Print(self):
        """ Time Complexity:O(n)"""
        temp = self.head  # O(1)
        while(temp): # o(n)
            print(temp.id,"->") # O(1)
            temp = temp.next # O(1)
            # O(1) + O(1) = O(1)
        # O(n) * O(1) = O(n)  
        # O(n) + O(1) = O(n)
            
    def pop(self):
        """ Time Complexity:O(n)"""
        if(self.head is self.tail):
            self.head = None # O(1)
            self.tail = None # O(1)
            # O(1) + O(1) = O(1)
            return
        temp = self.head # O(1)
        while(temp.next.next): # O(n)
            temp = temp.next # O(1)
            # O(n) * O (1) = O(n)
        self.tail = temp # O(1)
        self.tail.next = None # O(1)  
        # O(1) + O(n) + O(1) + O (n) = O(n)
        # O(1) or O(n) = O(n)
        
    def size(self):
        count = 0
        temp = self.head
        while(temp):
            count+=1
            temp = temp.next
        return (count)
stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
print("size of the stack is:",stack.size())
stack.Print()
print("pop One Element:")
stack.pop()
stack.Print()
print("size of the stack is:",stack.size())
print("pop One Element:")
stack.pop()
print("Top Element:",stack.top())
stack.Print()
print("pop One Element:")
stack.pop()
stack.Print()      
```

```python
#Stack implementation using Doubly Linked List
class NodeDLL:
    def __init__(self,val):
        self.id = val
        self.prev = None
        self.next = None
class StackDLL:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def push(self,val):
        """ Time Complexity:O(1)"""
        if(self.head is None):
            self.head = NodeDLL(val) # O(1)
            self.tail = self.head # O(1)
        else:
            temp = self.tail # O(1)
            self.tail.next = NodeDLL(val) # O(1)
            self.tail = self.tail.next # O(1)
            self.tail.prev = temp # O(1)
            # O(1) + O(1) + O(1) + O(1) = O(1)
        # O(1) or O(1) = O(1)
        
    def Print(self):
        """ Time Complexity:O(n)"""
        temp = self.head #O(1)
        while(temp): #O(n)
            print(temp.id) #O(1)
            temp = temp.next #O(1)
            #O(1) +O(1) = O(1)
        # O(n) * O(1) = O(n)
        # O(1) + O(n) = O(n)
            
    def pop(self):
        """ Time Complexity:O(1)"""
        if(self.head is self.tail):
            self.head = self.tail = None # O(1)
        else:
            self.tail = self.tail.prev # O(1)
            self.tail.next = None # O(1)
            # O(1) + O(1) = O(1)
        # O(1) or O(1) = O(1)
    
    def top(self):
        """ Time Complexity:O(n)"""
        return self.tail.id # O(1)
    
    def size(self):
        count = 0
        temp = self.head
        while(temp):
            count+=1
            temp = temp.next
        return (count)
        
def main():
    stack = StackDLL()
    stack.push(1)
    stack.push(2)
    stack.push(3)
    stack.Print()
    print("size of the stack is:",stack.size())
    
    print("Top element:",stack.top())
    print("After Pop:")
    stack.pop()
    stack.Print()
    print("size of the stack is:",stack.size())
    print("After Pop:")
    stack.pop()
    stack.Print()
    print("After Pop:")
    stack.pop()
    stack.Print()
    
main()           
        
        
        
```
