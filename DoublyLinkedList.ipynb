```python
#(1) Reverse a doublly linked list with stack
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

    def reverse(head):
        stack = []
        temp = head
        while temp is not None:
            stack.append(temp)
            temp = temp.next
        head = stack.pop()
        temp = head
        while stack:
            node = stack.pop()
            temp.next = node
            node.prev = temp
            temp = temp.next
            temp.next = None
        return head
#Test
head = Node(1)
head.next = Node(2)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next

head = Node.reverse(head)
while head is not None:
    print(head.data, end = " ")
    head = head.next
```

```python
#(2) Implement a data structure for a spin wheel like the one shown. The spin can be in ANY (clockwise / anticlockwise) direction. Players take turns to spin the wheel.

#Create Spin Wheel
class Node:
    def __init__(self, val):
        self.data = val
        self.next = None
        self.prev = None

class SpinWheel:
    def __init__(self):
        self.head = None
        self.tail = None

    def insert(self, val):
        temp = Node(val)
        if self.head is None:
            self.head = self.tail = temp
        else:
            self.tail.next = temp
            temp.prev = self.tail
            self.tail = temp
        return temp

SpinWheel = SpinWheel()
SpinWheel.insert(100)
SpinWheel.insert(750)
SpinWheel.insert(250)
SpinWheel.insert(10000)
SpinWheel.insert(100)
SpinWheel.insert(1500)
SpinWheel.insert(100)
SpinWheel.insert(2500)
SpinWheel.insert(500)
SpinWheel.insert(1500)
SpinWheel.insert(250)
SpinWheel.insert(5000)
SpinWheel.insert(750)
SpinWheel.insert(2500)
SpinWheel.insert(500)
temp = SpinWheel.insert(1000)
temp.next = SpinWheel.head
SpinWheel.head.prev = temp


#Generate Random Number
import random
RandomNumber = random.randint(1,10)

# Rotation.
print("1. Clockwise")
print("2. Anti Clockwise")
direction = int(input("Do you want to rotate the wheel in clockwise or anticlockwise rotation:"))
import time
if direction==1:
  # Spin the Wheel - clockwise
  temp = SpinWheel.head
  StartTime = time.time()
  while time.time() - StartTime < RandomNumber: # Changed startTime to StartTime
    print(temp.data,"")
    temp = temp.next
else:
  #Anti Cloclwise Rotation
  temp = SpinWheel.tail
  StartTime = time.time()
  # Spin the Wheel
  while time.time() - StartTime < RandomNumber: # Changed startTime to StartTime
    print(temp.data,"")
    temp = temp.prev

```
